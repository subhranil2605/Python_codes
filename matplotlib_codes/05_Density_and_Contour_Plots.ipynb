```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

# Density and Contour Plots

3D data in 2D using contours or color-coded regions.<br>
<code>plt.contour</code>: For contour plots<br>
<code>plt.contourf</code>: For filled contour plots<br>
<code>plt.imshow</code>: For images<br>

## Visualizing a Three-Dimensional Function

Contour plot using $z=f(x, y)$

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

<code>plt.contour</code> takes three args: a grid of x, y and z values<br>
The x and y values represent the positions on the plot, z value will be represented by the contour levels.<br>
<code>np.meshgrid</code> function builds 2D grids from 1D arrays:

```python
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black')
```

```python
plt.contour(X, Y, Z,30, cmap='Blues')   # Red-Gray

# plt.cm.<TAB> use this to get the colormaps
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='RdGy')
ax.set(xlim=(-6, 6), ylim=(-6, 6))
```
