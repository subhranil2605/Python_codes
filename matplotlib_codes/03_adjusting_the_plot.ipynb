```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 10, 1000)
```

# Adjusting the color

```python
'''
color by name, short color code, grayscale between 0 and 1, Hex code, rgb tyuple, HTML color names
'''
colors = ['blue', 'g', '0.75', '#FFDD44', (1.0, 0.2, 0.3), 'chartreuse']
for i in range(6):
    plt.plot(x, np.sin(x-i), color=colors[i])
plt.show()
```

# Line styles

```python
line_styles = ['solid', 'dashed', 'dashdot', 'dotted']

for i in range(len(line_styles)):
    plt.plot(x, x + i, linestyle=line_styles[i])
plt.show()
```

```python
line_styles = ['-', '--', '-.', ':']

for i in range(len(line_styles)):
    plt.plot(x, x + i, linestyle=line_styles[i])
plt.show()
```

# Combined linestyle and color

```python
line_style_with_color = [
    '-g',     # solied green
    '--c',    # dashed cyan
    '-.k',    # dot dashed black
    ':r'      # dotted red
]


for i in range(len(line_styles)):
    plt.plot(x, x + i, line_style_with_color[i])
plt.show()
```

# Adjusting the Plot: Axes Limits

Matplotlib automatically choose default axes limits for the plot. But if you want to customize it, use this: <code>plt.xlim()</code> and <code>plt.ylim()</code> 

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 4.5)

plt.show()
```

# Reverse the axes

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2)

plt.show()
```

To use in a single line code <code>plt.axis([x_low, x_high, y_low, y_high])</code>

```python
plt.plot(x, np.sin(x))

plt.axis([-1, 11, -1.2, 1.2])
```

<code>plt.axis('tight')</code>: Automatically tighten the bounds around the current plot

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

# Labeling Plots

Labeling the plots means: titles, axis, labels and simple legends

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel("x")
plt.ylabel("$sin(x)$")
plt.show()
```

Multiple Lines are being shwon within a single axes <code>plt.legend()</code>

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

plt.axis('equal')
plt.legend()
plt.title("A simple curve")
plt.xlabel("x")
plt.ylabel("$Sin(x)$ and $Cos(x)$")

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
ax.set_xlabel("x")
ax.set_ylabel("$Sin(x)$")
ax.set_xlim(-1, 11)
ax.set_ylim(-1.2, 1.2)
ax.set_title("A simple curve")

plt.show()
```

```python
ax = plt.axes()
ax.plot(x, np.cos(x))

ax.set(
    xlim=(0, 10),
    ylim=(-2, 2),
    xlabel="x",
    ylabel="$sin(x)$",
    title="A simple plot"
)

plt.show()
```
