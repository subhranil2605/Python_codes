```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
rng = np.random.RandomState(0)

for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label=f"marker={marker}")

plt.legend()
plt.show()
```

# Combining line and point markers

```python
plt.plot(x, y, '-ob')
```

```python
plt.plot(x, y, '-p')
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15,
         linewidth=4,
         markerfacecolor="white",
         markeredgecolor="gray",
         markeredgewidth=2
)
```

# Scatter Plots with plt.scatter

```python
plt.scatter(x, y, marker='d')
```

The primary difference of <code>plt.scatter()</code> and <code>plt.plot()</code> is that it can be used to create <br>
scatter plots where the properties of each individual point can be customized.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```
