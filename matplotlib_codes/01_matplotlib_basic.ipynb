```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

# Plotting from a script

```python
%matplotlib inline

import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
fig = plt.figure()

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

# Two Interfaces for the Price of One

Matplotlib has dual interfaces: <br>
(i) A convenient MATLAB-style state-based interface, <br>
(ii) Object-oriented interface

# MATLAB Style Interface

```python
plt.figure()      # create a plot figure

# create the first two panels and set the current axis
plt.subplot(2, 1, 1)  # (rows, columns, panel number)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.show()
```

# Object-oriented interface

```python
# first create a grid of plots

# ax is an array of two Axes objects
fig, ax = plt.subplots(2)

# call plot() method on the appropiate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
plt.show()
```
